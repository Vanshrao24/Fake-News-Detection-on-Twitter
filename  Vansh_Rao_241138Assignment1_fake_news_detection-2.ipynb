### Problem 1: Word Count Ignoring Punctuation (Built-in Functions)

**Task:** Write a Python function that counts the number of words in a given string, ignoring all punctuation. Only consider sequences of alphanumeric characters as words. You should only use Python's built-in string methods (no regex or NLP libraries).

```python
text = "Hello, world! How are you doing today? Let's count these words (properly)."

def word_count(text):
    # Replace punctuation with space, then split
    cleaned = ""
    for char in text:
        if char.isalnum() or char == " ":
            cleaned += char
        else:
            cleaned += " "
    words = [w for w in cleaned.split() if w]
    return len(words)

print(word_count(text))
```

### Problem 9: Custom Stop Word Removal

**Task:** Create a function that removes a predefined list of stop words from a given sentence. The function should return the cleaned sentence as a string.

```python
stop_words = ["is", "a", "the", "and", "with", "some"]
sentence = "This is a sample sentence with some stop words that need to be removed."

def remove_stop_words(sentence, stop_words):
    words = sentence.split()
    filtered = [w for w in words if w.lower() not in stop_words]
    return " ".join(filtered)

print(remove_stop_words(sentence, stop_words))
```

### Problem 11: Text Cleaning - Remove Numbers and Extra Spaces (Regex)

**Task:** Write a function that uses regular expressions to perform two cleaning steps on a text:
1.  Remove all digits (0-9).
2.  Replace multiple spaces with a single space.

The function should return the cleaned text.

```python
import re

text = "This text has 123 numbers and   extra   spaces, also 456 digits."

def clean_text(text):
    text = re.sub(r'\d+', '', text)       # Remove digits
    text = re.sub(r' +', ' ', text)       # Multiple spaces → single space
    return text.strip()

print(clean_text(text))
```

### Problem 12: Average Sentence Length (NLTK)

**Task:** Calculate the average word count per sentence in a given paragraph using NLTK's `sent_tokenize` and `word_tokenize`. Print the average length.

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')

paragraph = "NLP is amazing. It helps machines understand human language. This is a very interesting field of study and has many applications."

sentences = sent_tokenize(paragraph)
total_words = sum(len(word_tokenize(s)) for s in sentences)
avg_length = total_words / len(sentences)

print(f"Number of sentences: {len(sentences)}")
print(f"Average sentence length: {avg_length:.2f} words")
```

### Problem 13: Enhanced Rule-based Sentiment Analysis

**Task:** Modify the existing `sentiment_analysis` function to include a 'Mixed Sentiment' category. If the text contains both positive and negative words (and neither count is zero), it should be classified as 'Mixed Sentiment'. Otherwise, use the existing 'Positive', 'Negative', or 'Neutral' classifications.

```python
positive_words = {"happy", "good", "great", "excellent", "love", "amazing"}
negative_words = {"sad", "bad", "terrible", "hate", "awful", "worst"}

def sentiment_analysis_enhanced(text):
    words = text.lower().split()
    pos_count = sum(1 for w in words if w.strip(".,!?") in positive_words)
    neg_count = sum(1 for w in words if w.strip(".,!?") in negative_words)

    if pos_count > 0 and neg_count > 0:
        return "Mixed Sentiment"
    elif pos_count > 0:
        return "Positive"
    elif neg_count > 0:
        return "Negative"
    else:
        return "Neutral"

text1 = "I love this product, but the delivery was terrible."
text2 = "This is a great movie."
text3 = "The service was awful."
text4 = "It was okay."

print(f"Text 1 Sentiment: {sentiment_analysis_enhanced(text1)}")  # Mixed
print(f"Text 2 Sentiment: {sentiment_analysis_enhanced(text2)}")  # Positive
print(f"Text 3 Sentiment: {sentiment_analysis_enhanced(text3)}")  # Negative
print(f"Text 4 Sentiment: {sentiment_analysis_enhanced(text4)}")  # Neutral
```

**Task:** Apply `CountVectorizer` to the new corpus provided. Print the feature names (vocabulary) and the Bag-of-Words (BoW) representation (as an array) for each document in the corpus.

```python
from sklearn.feature_extraction.text import CountVectorizer

new_corpus = [
    "Machine learning is a fascinating field",
    "Deep learning is a subset of machine learning and uses neural networks",
    "Neural networks are powerful tools for AI"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(new_corpus)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```

### Problem 15: TF-IDF and Top 3 Keywords per Document

**Task:** Apply `TfidfVectorizer` to the provided corpus. Then, for each document, identify and print the top 3 most important keywords based on their TF-IDF scores.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

corpus_for_tfidf = [
    "The cat sat on the mat and purred happily",
    "A small dog ran in the park chasing a ball",
    "Cats and dogs are common pets and can be very playful"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus_for_tfidf)
feature_names = vectorizer.get_feature_names_out()

for i, row in enumerate(X.toarray()):
    top3_indices = np.argsort(row)[::-1][:3]
    top3_keywords = [(feature_names[j], round(row[j], 4)) for j in top3_indices]
    print(f"Doc {i+1}: {top3_keywords}")
```
