## Problem 1: Basic Tokenization

2. Tokenize the same sentence into words using `spaCy`.

## Problem 3: Stemming vs. Lemmatization

**Instructions:**
1. Apply NLTK's `PorterStemmer` to the provided word list.
2. Apply NLTK's `WordNetLemmatizer` (or spaCy's lemmatizer) to the same word list.
3. Print the results side-by-side to observe the differences.

**Words:**
`['running', 'flies', 'computed', 'better']`

```python

from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')

words = ['running', 'flies', 'computed', 'better']

# Stemming and Lemmatization loops here
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("word -> stem -> lemma")
for w in words:
    stem_word = stemmer.stem(w)
    lemma_word = lemmatizer.lemmatize(w)
    print(w, "->", stem_word, "->", lemma_word)
```

## Problem 4: Text Classification with CountVectorizer

**Instructions:**
1. Convert the small text dataset into a Bag-of-Words matrix using `CountVectorizer` from scikit-learn.
2. Train a `LogisticRegression` classifier on the data.
3. Predict the sentiment label (0 or 1) for the target review.

**Data:**
* `reviews = ["I loved this movie", "Horrible and boring", "An absolute masterpiece", "Waste of time"]`
* `labels = [1, 0, 1, 0]`
* `target_review = ["I loved the time spent on this movie"]`

```python

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

reviews = ["I loved this movie", "Horrible and boring", "An absolute masterpiece", "Waste of time"]
labels = [1, 0, 1, 0]
target_review = ["I loved the time spent on this movie"]

# Vectorize, fit model, and predict target here
vectorizer = CountVectorizer()
train_vectors = vectorizer.fit_transform(reviews)

model = LogisticRegression()
model.fit(train_vectors, labels)

target_vector = vectorizer.transform(target_review)
prediction = model.predict(target_vector)

print("Review:", target_review[0])
print("Predicted label (1 = positive, 0 = negative):", prediction[0])
```

## Problem 7: Regular Expressions for Pattern Extraction

**Instructions:**
1. Use Python's `re` module to locate and print all email addresses in the text.
2. Create a second regex pattern to locate and print all phone numbers formatting like `XXX-XXX-XXXX`.

**Text:**
`"Please contact support@company.com or reach out to sales@test.org. Alternatively, call us at 123-456-7890 or 555-888-9999."`

```python

import re

text = "Please contact support@company.com or reach out to sales@test.org. Alternatively, call us at 123-456-7890 or 555-888-9999."

# Find emails and phone numbers using re.findall()
emails = re.findall(r'[\w.]+@[\w.]+\.[a-zA-Z]+', text)
phones = re.findall(r'\d{3}-\d{3}-\d{4}', text)

print("Emails:", emails)
print("Phone numbers:", phones)
```

## Problem 9: Rule-Based Sentiment Analysis with VADER

**Instructions:**
1. Initialize NLTK's `SentimentIntensityAnalyzer` (VADER).
2. Check the polarity scores for both given sentences.
3. Print the **compound score** for each sentence and state whether it is overall positive or negative.

**Sentences:**
1. `"This new feature update is absolutely wonderful!"`
2. `"The customer support was horrible and the product broke instantly."`

```python

from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

s1 = "This new feature update is absolutely wonderful!"
s2 = "The customer support was horrible and the product broke instantly."

# Calculate and print scores here
sia = SentimentIntensityAnalyzer()

score1 = sia.polarity_scores(s1)
score2 = sia.polarity_scores(s2)

print("Sentence 1 compound score:", score1['compound'])
print("Sentence 1 overall:", "Positive" if score1['compound'] > 0 else "Negative")

print("Sentence 2 compound score:", score2['compound'])
print("Sentence 2 overall:", "Positive" if score2['compound'] > 0 else "Negative")
```

## Problem 11: Document Clustering with K-Means

**Instructions:**
1. Convert the list of 4 documents into numerical features using `TfidfVectorizer`.
2. Initialize and fit a `KMeans` clusterer with `n_clusters=2`.
3. Print the generated cluster labels corresponding to each text document to see how they grouped.

**Documents:**
`docs = ["Deep learning models train neural networks.", "Football players run on the pitch.", "Convolutional layers process visual imagery.", "The basketball team scored a three-pointer."]`

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

docs = ["Deep learning models train neural networks.",
        "Football players run on the pitch.",
        "Convolutional layers process visual imagery.",
        "The basketball team scored a three-pointer."]

# Extract TF-IDF features and cluster using KMeans here
vectorizer = TfidfVectorizer()
doc_vectors = vectorizer.fit_transform(docs)

km = KMeans(n_clusters=2, random_state=42)
km.fit(doc_vectors)

print("Cluster labels:", km.labels_)
for doc, label in zip(docs, km.labels_):
    print("Cluster", label, "-", doc)
```

**Instructions:**

## Problem 13: Rule-Based Aspect-Based Sentiment Analysis (ABSA)

**Instructions:**
1. Write a small Python script to check for two distinct aspects in the review below: `'screen'` and `'battery'`.
2. If the keyword `'screen'` appears, use a simple `if` condition to check if a positive word like `'beautiful'` or `'great'` is nearby. Print the derived aspect sentiment status.
3. Do the same for `'battery'` with negative terms like `'poor'` or `'bad'`.

**Review:**
`"The laptop has a beautiful screen but the battery performance is poor."`

```python

review = "The laptop has a beautiful screen but the battery performance is poor."

# Build simple keyword tracking logic to assign sentiment to 'screen' and 'battery'
review_lower = review.lower()

if 'screen' in review_lower:
    if 'beautiful' in review_lower or 'great' in review_lower:
        print("Screen sentiment: Positive")
    else:
        print("Screen sentiment: Negative")

if 'battery' in review_lower:
    if 'poor' in review_lower or 'bad' in review_lower:
        print("Battery sentiment: Negative")
    else:
        print("Battery sentiment: Positive")
```

## Problem 15: Frequency-Based Keyword Extraction

**Instructions:**
1. Split the lowercase version of the text into basic tokens.
2. Filter out basic English stop words.
3. Use NLTK's `FreqDist` to find the top 2 most frequent meaningful words (keywords) in the text, and print them.

**Text:**
`"Cloud computing infrastructure helps developers deploy code quickly. Cloud platforms minimize computing costs while maximizing global up-time scalability."`

```python

from nltk.probability import FreqDist
from nltk.corpus import stopwords

text = "Cloud computing infrastructure helps developers deploy code quickly. Cloud platforms minimize computing costs while maximizing global up-time scalability."

# Filter stop words out and find the 2 most common words here using FreqDist
words = text.lower().split()
stop_words = set(stopwords.words('english'))
clean_words = [w.strip('.,') for w in words if w.strip('.,') not in stop_words]

freq = FreqDist(clean_words)
top_keywords = freq.most_common(2)

print("Top 2 keywords:", top_keywords)
```
